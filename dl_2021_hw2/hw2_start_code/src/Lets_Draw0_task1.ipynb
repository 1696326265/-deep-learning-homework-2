```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
name = "Task1_modelResnet"

LA = np.load(name+"_Loss_Acc.npy")
```

```python
plt.title(name+' Loss')
plt.plot(LA[:,0], color='aqua', label='Train Loss')
plt.plot(LA[:,2], color='violet', label='Valid Loss')
plt.legend()
plt.show()
```

```python
plt.title(name+" Acc")
plt.plot(LA[:,1], color='aqua', label='Train Acc')
plt.plot(LA[:,3], color='violet', label='Valid Acc')
plt.legend()
plt.show()
```

```python
name2 = "Task2_modelResnet"

LA2 = np.load(name2+"_Loss_Acc.npy")
```

```python
plt.title(name+' vs '+name2+' Loss')
plt.plot(LA[:,0], color='aqua', label=name+' Train Loss')
plt.plot(LA[:,2], color='violet', label=name+' Valid Loss')
plt.plot(LA2[:,0], color='dodgerblue', label=name2+' Train Loss')
plt.plot(LA2[:,2], color='purple', label=name2+' Valid Loss')
plt.legend()
plt.show()
```

```python
plt.title(name+' vs '+name2+" Acc")
plt.plot(LA[:,1], color='aqua', label=name+' Train Acc')
plt.plot(LA[:,3], color='violet', label=name+' Valid Acc')
plt.plot(LA2[:,1], color='dodgerblue', label=name2+' Train Acc')
plt.plot(LA2[:,3], color='purple', label=name2+' Valid Acc')
plt.legend()
plt.show()
```

```python
shiftrate = 80/40
shiftrate

sft = []
for i in range(len(LA2)): sft.append(shiftrate*(i+0.5))
```

```python
plt.title(name+' vs '+name2+' Loss with Time-Normalized')
plt.plot(LA[:,0], color='aqua', label=name+' Train Loss')
plt.plot(LA[:,2], color='violet', label=name+' Valid Loss')
plt.plot(sft,LA2[:,0], color='dodgerblue', label=name2+' Train Loss')
plt.plot(sft,LA2[:,2], color='purple', label=name2+' Valid Loss')
plt.legend()
plt.show()
```

```python
plt.title(name+' vs '+name2+" Acc with Time-Normalized")
plt.plot(LA[:,1], color='aqua', label=name+' Train Acc')
plt.plot(LA[:,3], color='violet', label=name+' Valid Acc')
plt.plot(sft,LA2[:,1], color='dodgerblue', label=name2+' Train Acc')
plt.plot(sft,LA2[:,3], color='purple', label=name2+' Valid Acc')
plt.legend()
plt.show()
```
