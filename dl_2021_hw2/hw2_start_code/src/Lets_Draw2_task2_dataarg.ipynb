```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
name = "Task2_modelMyhighwaydataarg"

LA = np.load(name+"_Loss_Acc.npy")
```

```python
plt.title(name+' Loss')
plt.plot(LA[:,0], color='aqua', label='Train Loss')
plt.plot(LA[:,2], color='violet', label='Valid Loss')
plt.legend()
plt.show()
```

```python
plt.title(name+" Acc")
plt.plot(LA[:,1], color='aqua', label='Train Acc')
plt.plot(LA[:,3], color='violet', label='Valid Acc')
plt.legend()
plt.show()
```

```python
name2 = "Task2_modelMyhighway"

LA2 = np.load(name2+"_Loss_Acc.npy")
```

```python
plt.title(name+' vs '+name2+' Loss')
plt.plot(LA[:,0], color='aqua', label=name+' Train Loss')
plt.plot(LA[:,2], color='violet', label=name+' Valid Loss')
plt.plot(LA2[:,0], color='dodgerblue', label=name2+' Train Loss')
plt.plot(LA2[:,2], color='purple', label=name2+' Valid Loss')
plt.legend()
plt.xlabel("Epochs (x10)") 
plt.ylabel("Accuracy") 
plt.show()
```

```python
plt.title(name+' vs '+name2+" Acc")
plt.plot(LA[:,1], color='aqua', label=name+' Train Acc')
plt.plot(LA[:,3], color='violet', label=name+' Valid Acc')
plt.plot(LA2[:,1], color='dodgerblue', label=name2+' Train Acc')
plt.plot(LA2[:,3], color='purple', label=name2+' Valid Acc')
plt.legend()
plt.xlabel("Epochs (x10)") 
plt.ylabel("Accuracy") 
plt.show()
```

```python
shiftrate = 10/8
shiftrate

sft = []
for i in range(len(LA2)): sft.append(shiftrate*(i+0.5))
```

```python
plt.title(name+' vs '+name2+' Loss with Time-Normalized')
plt.plot(LA[:,0], color='aqua', label=name+' Train Loss')
plt.plot(LA[:,2], color='violet', label=name+' Valid Loss')
plt.plot(sft,LA2[:,0], color='dodgerblue', label=name2+' Train Loss')
plt.plot(sft,LA2[:,2], color='purple', label=name2+' Valid Loss')
plt.legend()
plt.show()
```

```python
# plt.title(name+' vs '+name2+" Acc")
plt.title("Babysited"+' vs '+"Origin"+" Acc")
idx = []
for i in range(len(LA)):
    idx.append(i)
p1 = 0
p2 = int(p1/shiftrate)
lim = 155
plt.plot(idx[p1:lim],LA[p1:lim,1], color='aqua', label='Babysited Train Acc')#name+' Train Acc')
plt.plot(idx[p1:lim],LA[p1:lim,3], color='violet', label='Babysited Valid Acc')#name+' Valid Acc')
plt.plot(sft[p2:],LA2[p2:,1], color='dodgerblue', label='Origin Train Acc')#name2+' Train Acc')
plt.plot(sft[p2:],LA2[p2:,3], color='purple', label='Origin Valid Acc')#name2+' Valid Acc')
plt.legend()
plt.xlabel("The number of cases trained (x8000)") 
plt.ylabel("Accuracy") 
# plt.scatter(40,0.8)
plt.show()
```
